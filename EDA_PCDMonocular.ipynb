```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import LabelEncoder
from pyquaternion import Quaternion
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
# Load the Excel file
df = pd.read_excel('PCD_ImageDepthFusion.xlsx')
```

```python
df_eda = df[['image_coord_x', 'image_coord_y', 'depth', 
'radar_x', 'radar_y']]
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

corr_matrix = df_eda.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm", square=True)
plt.title("Correlation Matrix of df_eda")
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set plot style
sns.set(style="whitegrid")

# Compute means
mean_radar_x = df_eda['radar_x'].mean()
mean_radar_y = df_eda['radar_y'].mean()
mean_depth = df_eda['depth'].mean()

# Create subplots
fig, axes = plt.subplots(1, 3, figsize=(14, 6))

# Histogram for radar_x
sns.histplot(df_eda['radar_x'], bins=50, kde=True, ax=axes[0], color='steelblue')
axes[0].axvline(mean_radar_x, color='black', linestyle='--', linewidth=1.5)
axes[0].text(mean_radar_x, axes[0].get_ylim()[1]*0.9, f'Mean: {mean_radar_x:.2f}', color='black', ha='center')
axes[0].set_title('Distribution of radar_x')
axes[0].set_xlabel('radar_x')
axes[0].set_ylabel('Frequency')

# Histogram for radar_y
sns.histplot(df_eda['radar_y'], bins=50, kde=True, ax=axes[1], color='darkorange')
axes[1].axvline(mean_radar_y, color='black', linestyle='--', linewidth=1.5)
axes[1].text(mean_radar_y, axes[1].get_ylim()[1]*0.9, f'Mean: {mean_radar_y:.2f}', color='black', ha='center')
axes[1].set_title('Distribution of radar_y')
axes[1].set_xlabel('radar_y')
axes[1].set_ylabel('Frequency')

# Histogram for depth
sns.histplot(df_eda['depth'], bins=50, kde=True, ax=axes[2], color='red')
axes[2].axvline(mean_depth, color='black', linestyle='--', linewidth=1.5)
axes[2].text(mean_depth, axes[2].get_ylim()[1]*0.9, f'Mean: {mean_depth:.2f}', color='black', ha='center')
axes[2].set_title('Distribution of depth')
axes[2].set_xlabel('depth')
axes[2].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set plot aesthetics
sns.set(style="ticks", color_codes=True)

# Generate pair plot
sns.pairplot(df_eda, diag_kind="kde", corner=True)

# Show the plot
plt.suptitle("Pair Plot of df_eda Features", y=1.02)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import seaborn as sns

# Set seaborn style for aesthetics
sns.set(style="whitegrid")

# Create 3D scatter plot
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

# Plot the data
ax.scatter(
    df_eda['radar_x'],
    df_eda['image_coord_y'],
    df_eda['depth']
)

# Label axes
ax.set_xlabel('Image Coord Y', fontsize=12)
ax.set_ylabel('Depth', fontsize=12)
ax.set_zlabel('Radar X', fontsize=12)

# Title and layout
ax.set_title('3D Scatter Plot: Image Coord Y vs Depth vs Radar X', fontsize=12)
plt.tight_layout()
plt.show()
```
